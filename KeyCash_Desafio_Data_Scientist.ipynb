## **Desafio Data Scientist**

#### **Premissas:**

* Os dados para resolução do desafio se encontram na planilha **Desafio_DS.xlsx**; 
* Possuímos uma operação de hotelaria onde locamos quartos por meio de operadoras terceiras como o Site A e Site B; 
* Cada operadora cobra uma taxa de serviço; 
* Somos responsáveis pela operação toda, desde o registro dos clientes na recepção até a distribuição e limpeza dos quartos; 
* Queremos expandir o nosso negócio para 200 unidades nos próximos dois meses (Dezembro/2019 e Janeiro/2020).

#### **Contexto:** 

O Mercado Hoteleiro se destaca dentro da cadeia produtiva do turismo como sendo um dos principais segmentos do setor em geração de receita e de empregabilidade.

**Seus principais meios de receita são:**
* As diárias de hospedagem, 
* Despesas com refeições e bares, 
* Locação de espaços para eventos e reuniões, 
* Lavanderia,
* Atividades de lazer.

**Seus principais custos dividem-se em:**
* Fixos (depreciação de imóveis e bens e salários administrativos),
* Variáveis (energia elétrica, água, alimentos, telefonia, lavanderia, materiais de limpeza e salários operacionais) 

**Principais Indicadores de desempenho:**
* Taxa de ocupação: Indica a porcentagem de unidades ocupadas em relação ao total disponível.

* Diária média: Trata-se do preço médio calculado para cada diária vendida em um determinado período, ou a receita total de hospedagens dividido pelo número total de diárias vendidas.

* RevPAR:  Receita por apartamento disponível (Revenue per Avaliable Room). Mostra a receita média obtida por cada quarto disponível. 

  É calculado pela divisão da receita de apartamentos pelo total de apartamentos disponíveis no ano, ou multiplicando diretamente a taxa de ocupação anual pela diária média.

**Desafios operacionais da expansão:**

* Apartamentos: Um número maior de quartos acarreta aumento em salários e encargos de recepção, governança, reservas e limpeza. Como também em material de limpeza, lavanderia, comissão aos agentes de viagens, tv a cabo.

* Alimentos e Bebidas: Custos diretos dos alimentos, salário e encargos da cozinha, restaurante, bar e room service. 
Como também louças, talheres, materiais para cozinha e restaurante.

* Administração: Além de salários e encargos do departamento de administração incluindo a gerência, contabilidade, recursos humanos e controladoria, podendo incluir também segurança, lazer, informática, sistemas e material de escritório.

* Manutenção: salários e encargos, despesas com manutenção de equipamentos, instalações, móveis e piscina.

**Solução:**

* Planejamento de demanda e Negociação com fornecedores: 

  * Montar um calendário com cronograma de ações a serem realizadas e previsão de demanda dão suporte para melhor execução operacional.
Custos com equipamentos, materiais diversos e alimentos e bebidas se negociados com antecedencia é possível ter ganhos em escala, redução de custos e prazos melhores.

  * Negociar o prazo de entrega para alguns materiais, de forma que cheguem nas épocas definidas, evita-se acúmulo de material quando não necessário (Ex: na reforma de quartos não adianta ter enxoval enquanto o quarto não estiver pronto, mas a compra pode ser feita em lote para ganhar desconto) ou evitando desperdício de comida em épocas de baixa temporada (negociar com produtores uma demanda prévia, mas garantir a entrega desses produtos no agendamento realizado).
* Treinamento de funcionários e melhoria de processos:
  * Ter uma equipe capacitada e que receba treinamentos constantes permite ter um atendimento padronizado e que proporcione melhor experiência para seus clientes (ex: atendentes que possam atender os clientes estrangeiros).
  * Obter feedback das equipes para identificar processos atuais que estão sendo realizados e que podem ser automatizados, principalmente na parte administrativa em busca de melhorar e aumentar produtividade. 

* Manutenção preventiva:
  * Fazer o planejamento e manutenção preventiva de equipamentos e instalações para se evitar o máximo de ter problemas na época de maior demanda.

## Baseline - Persistence Algorithm

```
from pandas import datetime
from pandas import DataFrame
from pandas import concat
from matplotlib import pyplot
from sklearn.metrics import mean_squared_error
 
# Criando o dataset baseline
df_baseline = df_base.copy()
df_baseline.head()
```

```
# Criando o lagged dataset das observações atuais em t e a previsão em t+1 (usando os dados de t como previsão)
# values = pd.DataFrame(df_baseline.values)

values = pd.DataFrame(df_baseline.loc[:,'receita_bruta_estimada'])
dataframe = concat([values.shift(1), values], axis=1)
dataframe.columns = ['t', 't+1']
print(dataframe.head(5))
```

```
# Separar em dados de treino e teste, e excluir a 1a linha pois apresenta valor NaN
X = dataframe.values
train_size = int(len(X) * 0.66)
train, test = X[1:train_size], X[train_size:]
train_X, train_y = train[:,0], train[:,1]
test_X, test_y = test[:,0], test[:,1]
 
# Criando o persistence model
def model_persistence(x):
	return x
 
# Criando modelo de validação
predictions = list()
for x in test_X:
	yhat = model_persistence(x)
	predictions.append(yhat)
test_score = mean_squared_error(test_y, predictions)
print('Test MSE: %.3f \n' % test_score)
 
# Plotar as previsões e resultados esperados
plt.figure(figsize=(15,12))
pyplot.plot(train_y)
pyplot.plot([None for i in train_y] + [x for x in test_y])
pyplot.plot([None for i in train_y] + [x for x in predictions])
pyplot.show()
```

```
# Importando as libraries
import warnings
warnings.filterwarnings("ignore")
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from pandas.plotting import autocorrelation_plot
from pandas import datetime
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
import numpy as np



```

```
# Criando o dataset com index em reservado
series = pd.read_excel('Desafio_DS.xlsx', header=0, usecols= ['Reservado', 'Receita bruta estimada'], parse_dates=[0], index_col=0, squeeze=True)

#series = series['receita_bruta_estimada']
#series = df_base.copy()
series = series.groupby(by='Reservado').agg({'receita_bruta_estimada': 'sum'})
print(series.head())
print('\n')

# Gerando o gráfico da série temporal
plt.figure(figsize=(20,5))
plt.plot(series)
plt.show()
```

O gráfico acima mostra uma tendência de crescimento ao longo do tempo.
Isso sugere que a série temporal não é estacionária.
Vamos fazer a diferenciação para verificar se a série temporal é estacionária ou não

```
# Plotar o gráfico de autocorrelação para confirmar a série estacionária
plot_acf(series)
plt.show()
```

O gráfico de autocorrelação mostra uma correlação positiva entre os dados indicando que o componente AR deve ser priorizado no modelo ARIMA


---



## Diferenciação

```
# Criando a 1a diferenciação e verificar se a série se torna estacionária
series_diff1 = series.diff().fillna(series)

plt.figure(figsize=(20,5))
plt.plot(series_diff1)
plt.show()
```

Fazendo a diferenciação vemos que a série temporal tem média e variância constante, ou seja, a série temporal se tornou estacionária.


---



```
# Plotar os gráficos ACF e PACF para confirmar a série estacionária
plot_acf(series_diff1)
plot_pacf(series_diff1)
plt.show()
```

Nos gráficos de ACF e PACF vemos que após a 1a diferenciação, os dados decresceram de forma rápida, mostrando que estão estacionários e de forma positiva, ou seja, iremos usar mais do componente AR do modelo ARIMA

## Fazendo as previsões no teste

```
history = [x for x in train]
predictions = list()

for t in range(len(test)):
	model = ARIMA(history, order=(3,1,1)).fit(disp=0)
	output = model.forecast()[0]
	predictions.append(output)
	obs = test[t]
	history.append(obs)
	#print('predicted=%f, expected=%f' % (yhat, obs))

# Verificando o % de erro no modelo entre os dados de teste e os valores previstos 
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)

# Plot comparando os valores de teste e valores previstos
plt.plot(test)
plt.plot(predictions, color='red')
plt.show()
```
